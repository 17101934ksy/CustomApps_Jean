```python
from urllib.request import urlopen, Request
import json
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import re
from time import sleep
import random
import time
import traceback



def ErrorLog(error: str): 
    current_time = time.strftime("%Y.%m.%d/%H:%M:%S", time.localtime(time.time())) 
    with open("Log.txt", "a") as f: 
        f.write(f"[{current_time}] - {error}\n")

start = time.time()
pagenum = 6
breaknum = 0
jean_list = []

# 상품번호
file_name = "jean_nums.txt"
file_open = open(file_name, "r", encoding="utf-8")
a=''
# 상품번호
for i in file_open:
    a += i
a=a.split(' ')
for num in a:
    jean_list.append({'num':num})
file_open.close()
for page in range(1,7):
    if 10<len(jean_list):
        print(num)
        break;
    else:
        if ((page % 2) == 1):
            sleep(random.uniform(8, 9))
            print("breaking")
        url = "https://search.musinsa.com/category/003002?&page="+str(pagenum+1)
        res = requests.get(url)
        res.raise_for_status()
        soup = BeautifulSoup(res.text, "lxml")
        r = re.compile('#pol([0-9]+)')
        jeans = soup.find_all("a", attrs={"href": r})
        for jean in jeans:
            num = ''.join(r.findall(jean['href']))
            jean_list.append({'num': num})
            if 10<len(jean_list):
                print(num)
                break;
            else:
                file_open.write(num+' ')
        pagenum += 1
        print(str(pagenum)+"페이지 입니다")

# 상품번호로 상품 정보 가져오기 시작
try:
    for jean in jean_list:
        breaknum += 1
        if(breaknum % 10 == 0):
            sleep(random.uniform(12, 15))
            print(str(jean)+"breaking")

        url = "https://store.musinsa.com/app/goods/"+str(jean['num'])
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36'+
            ' (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
        res = requests.get(url, headers=headers)
        soup = BeautifulSoup(res.text, "lxml")

        # 상품 이름
        name = soup.select('span.product_title em')
        a = re.sub('<.+?>', '', str(name), 0, 0).strip()
        name = a[1:-1]
        jean.update(name=name)

        # 상품 브랜드
        brand = soup.select('p.product_article_contents strong a')
        a = re.sub('<.+?>', '', str(brand), 0, 0).strip()
        brand = a[1:-1]
        jean.update(brand=brand)

        # 상품 해시태그
        tag = soup.select('p.product_article_contents a.listItem')
        a = re.sub('<.+?>', '', str(tag), 0, 0).strip()
        tag = re.sub(',', '',a, 0, 0).strip().replace(" ", '')
        tag = tag[1:-1]
        jean.update(tag=tag)

        # 상품 별점
        rate = soup.select('span.prd-score__rating')
        a = re.sub('<.+?>', '', str(rate), 0, 0).strip()
        rate = a[1:-1]
        jean.update(rate=rate)

        # 상품 이미지 주소
        img = str(soup.select('div.product-img img#bigimg'))
        a = (img.find('src'))+5
        b = (img.find('title'))-2
        img = "https:"+img[a:b]
        jean.update(img=img)
        
        try:
            xs= soup.select('tbody tr td.goods_size_val')[4]
            a = re.sub('<.+?>', '', str(xs), 0,0).strip()
            jean.update(xs=a)
        except:
            jean.update(xs='null')
        
        # 상품 구매 횟수
        headers = {
            'referer': url,
            'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36'+
            '(KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}

        url = 'https://m.store.musinsa.com/app/product/goodsview_stats/' +str(jean['num'])+'/0?&menu=view'
        
        response = urlopen(Request(url, headers=headers)).read().decode('utf-8')
        tp = str(json.loads(response)['purchase']['total_categories'])
        ap = str(json.loads(response)['purchase']['categories'])
        tv = str(json.loads(response)['pageview']['total_categories'])
        av = str(json.loads(response)['pageview']['categories'])

        jean.update(tp=tp)
        jean.update(tv=tv)
        jean.update(ap=ap)
        jean.update(av=av)
        
        url = 'https://m.store.musinsa.com/app/svc/member_price_new/'+str(jean['num'])+'/0?1633576324299&type=detail&skin_type=v2'
        response = urlopen(Request(url, headers=headers)).read().decode('utf-8')
        soup = BeautifulSoup(response, "lxml")
        a = int(re.sub(r'[^0-9]', '', str(soup.select('span.price')[1]),0,0))     #회원가
        jean.update(price=a)
        print(jean['num'],"완료")
        
        
    sleep(random.uniform(10, 15))   
        
    new_int_jean_nums = []

    for jean in jean_list:
        jean_num = int(jean["num"])
        new_int_jean_nums.append(jean_num)
    
    temp1,temp2,temp3,temp4,temp5,temp6,temp7,temp8,temp9,temp10,temp11,temp12 =[],[],[],[],[],[],[],[],[],[],[],[]
    for i in new_int_jean_nums:
        if(len(temp1)<=50):
            temp1.append(i)
        elif(len(temp2)<=50):
            temp2.append(i)
        elif(len(temp3)<=50):
            temp3.append(i)
        elif(len(temp4)<=50):
            temp4.append(i)
        elif(len(temp5)<=50):
            temp5.append(i)
        elif(len(temp6)<=50):
            temp6.append(i)
        elif(len(temp7)<=50):
            temp7.append(i)
        elif(len(temp8)<=50):
            temp8.append(i)
        elif(len(temp9)<=50):
            temp9.append(i)
        elif(len(temp10)<=50):
            temp10.append(i)
        
    temp11.append(temp1)
    temp11.append(temp2)
    temp11.append(temp3)
    temp11.append(temp4)
    temp11.append(temp5)
    temp11.append(temp6)
    temp11.append(temp7)
    temp11.append(temp8)
    temp11.append(temp9)
    temp11.append(temp10)
        
        
    headers = {
    'content-type': 'application/json',
    'user-agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N)'+
    'AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.71 Mobile Safari/537.36'}
        
    for i in temp11:
        data = '{"relationIds":'+str(i)+'}'

        response = requests.post(
            'https://like.musinsa.com/like/api/v2/liketypes/goods/counts', headers=headers, data=data)

        counttext = response.text.replace('"count":', 'count').split(',')
        counttemp = [word for word in counttext if'count' in word]
        countheart = re.sub('count', '', str(counttemp), 0, 0)[1:-1]
        heart = re.sub(r"[^,0-9]", "", countheart).split(',')
        
        for i in heart:
            temp12.append(i)
            
    for jean,j in zip(jean_list, temp12):
        try:
            jean.update(heart=j)
        except:
            jean.update(heart=0)
            print(jean['num'])
            print("오류로 인해 좋아요에 0을 입력하였습니다, 계속 진행 중 취소하지 마세요.")

    # json 데이터 출력
    with open('musinsa_jeans.json', 'w', encoding="utf-8") as f:
        f.write(json.dumps(jean_list,ensure_ascii=False))

    print(time.time() - start)
    print("완료")
    
    
except:
    print("에러발생 - 작업 저장 중")
    with open('musinsa_jeans.json', 'w', encoding="utf-8") as f:
        f.write(json.dumps(jean_list,ensure_ascii=False))
    
    err = traceback.format_exc() 
    ErrorLog(str(err))            
    print(time.time() - start)
```
