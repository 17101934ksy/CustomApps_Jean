```python
import re
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
import time
import random
from time import sleep
import json
import traceback


start = time.time()

def ErrorLog(error: str): 
    current_time = time.strftime("%Y.%m.%d/%H:%M:%S", time.localtime(time.time())) 
    with open("Log1txt", "a") as f: 
        f.write(f"[{current_time}] - {error}\n")


# jean_nums txt파일로 저장한 것 불러오기
f = open('./jean_nums.txt', 'r')
jean_nums = f.read().split()
f.close()

review = []
file_name = "review.json"
file_open = open(file_name, "a+", encoding="utf-8")


try:
    for jean_num in jean_nums:
        for i in range(10):
            print(str(jean_num)+"의"+str(i)+"페이지 cw")
            url = "https://store.musinsa.com/app/reviews/goods_estimate_list/style/"+jean_num+"/0/" +str(i+1)+"?similar_no=986708&select_similar_no=&is_cache=N&sort=new"
            headers = {
                'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36'+
                '(KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
            res = requests.get(url, headers=headers)
            res.raise_for_status()
            if(i%2==1):
                sleep(random.uniform(8, 9))
                print(str(i)+"번째 breaking")
            soup = BeautifulSoup(res.text, "html.parser")  # lxml

            for i in range(10):
                body = soup.select('div.review-list p.review-profile__body_information')[i]
                body = re.sub('<.+?>', '', str(body), 0, 0).strip().replace(" ", "")
                a=body.split(',')
                gender=a[0]
                height=float(a[1][:-2])
                weight=float(a[2][:-2])
                
                # print(body)
                name = soup.select('div.review-list a.review-goods-information__name')[i]
                name = re.sub('<.+?>', '', str(name), 0, 0).strip().replace(" ", "")
                # print(name)
                size = soup.select('div.review-list span.review-goods-information__option')[i]
                size = re.sub('<.+?>', '', str(size), 0, 0).strip().replace(" ", "")
                a = size.find(',')
                if(a != -1):
                    size = size[:a-1]
                else:
                    size
                    
                a = str(soup.select('div.review-list span.review-list__rating__unit')[i])
                #print(a)
                b= a.find('width: ')
                c= a.find('"></span>')
                rating = float("{:1f}".format(float(str(a[b+7:c])[:-1])/20))
                
                evl = soup.select('div.review-list ul.review-evaluation__list')[i]
                evl = re.sub('<.+?>', '', str(evl), 0, 0).strip().replace(" ", "").split('\n')
                evls=int({len(evl[0])==5 : 3, len(evl[0])==8: 2 }.get(True,1))
                evlb=int({len(evl[1])==5 : 3, len(evl[1])==7: 2 }.get(True,1))
                evlc=int({len(evl[2])==6 : 3, len(evl[2])==7: 2 }.get(True,1))
                evlt=int({len(evl[3])==7 : 3, len(evl[3])==8: 2 }.get(True,1))
                
                # print(evl)
                link = str(soup.select('div.review-list li.review-content-photo__item img')[i])
                a = int(link.find("src"))+5
                b = -3
                link = link[a:b]
                review.append({"num": jean_num, "gender":gender, "height":height,"weight":weight, "name": name, "rating": rating, "size": size, "evls": evls,"evlb": evlb, "evlc": evlc, "evlt": evlt, "link": link})
                
    with open('review.json', 'w', encoding="utf-8") as f:
        f.write(json.dumps(review,ensure_ascii=False))
        
except:
    print("에러발생 - 작업 저장 중")
    with open('review.json', 'w', encoding="utf-8") as f:
        f.write(json.dumps(review,ensure_ascii=False))
    
    err = traceback.format_exc() 
    ErrorLog(str(err))            
    print(time.time() - start)
```
